```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv(r"D:\house_price_prediction\data\raw\train.csv")
```

```python
x_train = pd.read_csv(r"D:\house_price_prediction\data\processed\x_train.csv")
x_test = pd.read_csv(r"D:\house_price_prediction\data\processed\x_test.csv")
y_train = pd.read_csv(r"D:\house_price_prediction\data\processed\y_train.csv")
y_test = pd.read_csv(r"D:\house_price_prediction\data\processed\y_test.csv")
```

```python
y_train=y_train.squeeze()
y_test=y_test.squeeze()
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred))
r2_lr=r2_score(y_test,y_pred)
print(f"{rmse_lr:.4f}")
print(f"{r2_lr:.4f}")
```

```python
la=Lasso(alpha=1.0)
la.fit(x_train, y_train)
```

```python
y_pred_la = la.predict(x_test)
rmse_la = np.sqrt(mean_squared_error(y_test, y_pred_la))
r2_la = r2_score(y_test, y_pred_la)
```

```python
print(f"{rmse_la:.4f}")
print(f"{r2_la:.4f}")
```

```python
re = Ridge(alpha=1.0)
re.fit(x_train, y_train)
```

```python
y_pred_re = re.predict(x_test)
rmse_re = np.sqrt(mean_squared_error(y_test, y_pred_re))
r2_re = r2_score(y_test, y_pred_re)
```

```python
print(f"{rmse_re:.4f}")
print(f"{r2_re:.4f}")
```

```python
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
print(f"{rmse_rf:.4f}")
print(f"{r2_rf:.4f}")
```

```python
gbr = GradientBoostingRegressor(n_estimators=200, random_state=42)
gbr.fit(x_train, y_train)
```

```python
y_pred_gbr = gbr.predict(x_test)
rmse_gbr = np.sqrt(mean_squared_error(y_test, y_pred_gbr))
r2_gbr = r2_score(y_test, y_pred_gbr)
```

```python
print(f"{rmse_gbr:.4f}")
print(f"{r2_gbr:.4f}")
```

```python
import pickle

# Save the trained Gradient Boosting model
with open("model.pkl", "wb") as f:
    pickle.dump(gbr, f)

print("Gradient Boosting model saved successfully as 'model.pkl'")
```
