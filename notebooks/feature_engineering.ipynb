```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"D:\house_price_prediction\data\raw\train.csv")
```

```python
drop_cols = ["PoolQC", "MiscFeature", "Alley", "Fence"]
df.drop(columns=drop_cols,inplace=True)
```

```python
df.drop(columns=["Id"], inplace=True)
```

```python
num_cols=df.select_dtypes(include=["int64","float64"]).columns
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols= df.select_dtypes(include=["object"]).columns
df[cat_cols]=df[cat_cols].fillna("None")
```

```python
df["SalePrice"] = np.log(df["SalePrice"])
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
x=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.to_csv(r"D:\house_price_prediction\data\processed\x_train.csv", index=False)
x_test.to_csv(r"D:\house_price_prediction\data\processed\x_test.csv", index=False)
y_train.to_csv(r"D:\house_price_prediction\data\processed\y_train.csv", index=False)
y_test.to_csv(r"D:\house_price_prediction\data\processed\y_test.csv", index=False)
```
